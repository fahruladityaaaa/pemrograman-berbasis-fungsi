```python
# Menggunakan prosedural programming 
#implementasi fungsi komposisi dengan nilai dan tiga fungsi dalam pemrograman prosedural
def f(x):
    return 4 ** x
print(f"hasil dari f(x)= 4**x dengan x adalah 3 = {f(3)}")

def g(x):
    return x ** 3
print(f"hasil dari g(x)= x**x dengan x adalah = {g(3)}")

def h(x):
    return 3 + x
print(f"hasil dari h(x)= 3+x dengan x adalah 3 = {h(3)}")

def komposisi_nilai(x):
    return h(g(f(x)))

print(f"nilai fungsi komposisinya adalah = {komposisi_nilai(3)}")
```

```python
# Menggunakan functional programming


fx= lambda x: 4**x
print(f"hasil dari f(x)= 4**x dengan x adalah 3 = {fx(3)}")

gx= lambda x: x**3
print(f"hasil dari g(x)= X**x dengan x adalah 3 = {gx(3)}")

hx= lambda x: 3 + x
print(f"hasil dari h(x)= 3+x dengan x adalah 3 = {hx(3)}")

komposisi_lambda = lambda x: hx(gx(fx(x)))
print(f"nilai fungsi komposisinya adalah = {komposisi_lambda(3)}")
```

```python
# Menggunakan object oriented programming
class Fungsi:
  def _init_(self):
    pass

  def f(self, x):
    return 4 ** x

  def g(self, x):
    return x ** 3

  def h(self, x):
    return 3 + x

  def komposisi(self, x):
    return self.h(self.g(self.f(x)))


fungsi = Fungsi()
print(f"hasil dari f(x)= 4**x dengan x adalah 3 = {fungsi.f(3)}")
print(f"hasil dari g(x)= x**x dengan x adalah 3 = {fungsi.g(3)}")
print(f"hasil dari h(x)= 3+x dengan x adalah 3 =  {fungsi.h(3)}")
print(f"nilai fungsi komposisinya adalah = {fungsi.komposisi(3)}")
```

##Pemrograman Prosedural

```python
def bisection_method_procedural(f, a, b, tol):
    # Mendefinisikan fungsi f(x) = x - 2**(-x)
    def f(x):
        return x - 2**(-x)

    # Memilih tebakan awal x0 dan x1 sehingga f(x0)f(x1) < 0
    if f(a) * f(b) >= 0:
        print("Tebakan awal tidak valid")
        return None

    # Menentukan nilai batas awal a dan b
    x0 = a
    x1 = b
    x2 = (x0 + x1) / 2

    # Mengatur Perhitungan ke 0
    counter = 0

    # Mengulangi nilai sampai |f(x2)| <= tol atau loket > 1000
    while abs(f(x2)) > tol and counter < 1000:

        # Menghitung f(x0)f(x2)
        fx0x2 = f(x0) * f(x2)

        # Memperbarui x0 dan x1 berdasarkan nilai f(x0)f(x2)
        if fx0x2 < 0:
            x1 = x2
        elif fx0x2 > 0:
            x0 = x2
        else:
            break

        # Menghitung Pembaruan x2
        x2 = (x0 + x1) / 2

        # Kita tingkatkan penghitungan 
        counter += 1

    # Setelah itu kita cetak hasilnya
    if counter >= 1000:
        print("Metode gagal konvergen")
    else:
        print(f"Akar yang dihitung adalah {x2:.6f}")

# dan berikut adalah contoh dari bentuk pemanggilan
a = 0.1
b = 0.9
tol = 1e-5
bisection_method_procedural(f, a, b, tol)
```

##Pemrograman Fungsional

```python
def bisection_method_functional(f, a, b, tol):
    # Mendefinisikan fungsi f(x) = x - 2**(-x)
    def f(x):
        return x - 2**(-x)

    # Memilih tebakan awal x0 dan x1 sehingga f(x0)f(x1) < 0
    if f(a) * f(b) >= 0:
        print("Tebakan awal tidak valid")
        return None

    # Lalu kita tentukan fungsi rekursif untuk melakukan metode bagi dua
    def bisection(f, a, b, tol, counter):
        x0 = a
        x1 = b
        x2 = (x0 + x1) / 2
        fx0x2 = f(x0) * f(x2)
        if abs(f(x2)) <= tol or counter >= 1000:
            return x2
        elif fx0x2 < 0:
            return bisection(f, x0, x2, tol, counter + 1)
        else:
            return bisection(f, x2, x1, tol, counter + 1)

    # Kemudian kita panggil fungsi rekursif dan cetak hasilnya
    result = bisection(f, a, b, tol, 0)
    if result is None:
        print("Metode gagal konvergen")
    else:
        print(f"Akar yang dihitung adalah {result:.6f}")

# Dan berikut adalah bentuk contoh pemanggilan fungsinya
a = 0.1
b = 0.9
tol = 1e-5
bisection_method_functional(f, a, b, tol)
```
